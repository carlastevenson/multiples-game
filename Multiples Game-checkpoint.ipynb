```python
from ipycanvas import Canvas, MultiCanvas, hold_canvas
import ipywidgets as widgets
from ipywidgets import Button, Output
from IPython.display import display
import numpy as np
import random
```

```python
print("This game generates 40 random numbers which include at least 15 numbers of the multiple that you choose.")
number=input("Enter the multiple:")

#Generate multiples of the number entered (int_list) and additional numbers (filler_numbers) for the display (board_entries).
number_of_correct_answers=15
int_list=[int(number)*(n+1) for n in range(number_of_correct_answers)]
random.shuffle(int_list)
filler_numbers=[random.randrange(1,10*int(number)+1,1) for m in (range(25))]

board_entries=int_list+filler_numbers
random.shuffle(board_entries)

#print(board_entries)

#Set location of entries (x,y).
mcanvas=MultiCanvas(3,width=800,height=500)
mcanvas[0].font="48px serif"
x=20
y=60
space=100
answer_location_dict={}
correct_prob_numbers=[]
with hold_canvas():
    for n in range(5):
        for m in range(8):
            mcanvas[0].fill_text(str(board_entries[m+8*n]),x+m*space,y+n*space)
            answer_location_dict[m+8*n]=(x+m*space,y+n*space)
            
            #Create a dictionary of problem numbers (in base 8) for correct answers.  
            if board_entries[m+8*n]%int(number)==0:
                correct_prob_numbers.append(m+8*n)
            mcanvas[0]


    
```

```python
#Create a list of problem numbers selected by the student.
student_ans_list=[]

out=widgets.Output()
@out.capture()

def place_point(x,y):
    global x_pt, y_pt, x0, y0, student_ans_list
    x_pt=x
    y_pt=y
    
    m=round((x_pt-20)/100)
    n=round((y_pt-20)/100)
   
    
    x0=m*space
    y0=n*space
   
    
    
    #Create dictionary of problem numbers (in base 8) for student answers.
    student_ans_list.append(m+8*n)
    mcanvas[1].fill_style="white"
    mcanvas[1].fill_rect(x0,y0,space,space)
    
    return x_pt,y_pt,x0,y0,student_ans_list

display(out)          #This assures that the output of the function is displayed on the canvas.
mcanvas.on_mouse_down(place_point)

def check_answers(object):
    student_ans_list.sort()
    
    mcanvas[2].fill_style="red"
    mcanvas[2].font="36px serif"
    
    for number in student_ans_list:
        x=answer_location_dict[number][0]
        y=answer_location_dict[number][1]
        
        if number in correct_prob_numbers:
            mcanvas[2].fill_text("\u2713",x,y)
        else:
            mcanvas[2].fill_text("X",x,y)
    mcanvas[2]

def correct_answers(object):
    #Make all numbers visible again.
    mcanvas[1].clear()
    mcanvas[0].stroke_style="red"
    mcanvas[0].line_width=2
    for number in correct_prob_numbers:
        x=answer_location_dict[number][0]+5
        y=answer_location_dict[number][1]-40
        
        mcanvas[0].stroke_circle(x+20,y+20,45)
    mcanvas[0]
        
    
btn1=widgets.Button(description="Check Answers")
btn1.on_click(check_answers)

btn2=widgets.Button(description="Correct Answers")
btn2.on_click(correct_answers)

ui=widgets.HBox([btn1,btn2])
display(ui)

mcanvas
```
